# SEER Phase 2 generation on Colab A100
Expensive and network-capable toggles default to `False`. Run cells in order.

```python
import torch
from google.colab import drive

drive.mount('/content/drive')
assert torch.cuda.is_available(), 'Select a GPU runtime'
name = torch.cuda.get_device_name(0)
assert 'A100' in name, f'A100 required, found {name}'
print(name)
```

```python
# Configure either a clone or an existing Drive/uploaded repository.
REPO_URL = 'https://github.com/OWNER/seer.git'
REPO_BRANCH = 'main'
REPO_DIR = '/content/seer'
USE_EXISTING_REPO = False
PREPARED_ROOT = '/content/drive/MyDrive/seer/phase02-prepared-v2'
RUN_MODEL_DOWNLOAD = False
RUN_PRIMARY = False
RUN_THINKING = False
```

```python
import pathlib
import subprocess

if not USE_EXISTING_REPO and not pathlib.Path(REPO_DIR).exists():
    subprocess.run(['git','clone','--branch',REPO_BRANCH,'--depth','1',REPO_URL,REPO_DIR],check=True)
subprocess.run(['pip','install','uv'],check=True)
```

```python
# Validation always runs; model download and generation remain explicit.
cmd = ['python', f'{REPO_DIR}/scripts/colab_phase2_generation.py',
       '--repo-dir', REPO_DIR, '--prepared-root', PREPARED_ROOT]
if RUN_MODEL_DOWNLOAD:
    cmd += ['--allow-model-download']
if RUN_PRIMARY:
    cmd += ['--run-primary']
if RUN_THINKING:
    cmd += ['--run-thinking']
subprocess.run(cmd, check=True)
```

## Durable artifacts and estimates
